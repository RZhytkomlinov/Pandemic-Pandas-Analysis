```python
copyset = children_dataset.copy()
```

```python
group_one_c = copyset[copyset.Group == 1]
```

```python
group_two_c = copyset[copyset.Group == 2]
```

```python
group_one_c['reg_freq'] = group_one_c.groupby('Region').cumcount() >= 20
```

```python
group_one_c[group_one_c.reg_freq == True].describe()
```

```python
group_two_c['reg_freq'] = group_one_c.groupby('Region').cumcount() >= 20
```

```python
group_two_c[group_two_c.reg_freq == True].describe()
```

Regions that appeared frequently in first group aren't that frequent in the second one hence we can't make any assumptions based on regions
